```python
tas_ground = pd.read_csv("../ICESAT/tasmania/tasmania_g.csv")
minimum = tas_ground['g_height'].mean()
tas_ground

relative_ht_min = [i - minimum for i in tas_ground['g_height']]

tas_ground['relative_ht'] = relative_ht_min
tas_ground = tas_ground.drop(columns=['Unnamed: 0','g_height'], axis=1)
tas_ground.rename(columns={"relative_ht": "g_height"}, inplace=True)


tas_ground.to_csv('../ICESAT/tasmania/tasmania_g_mean.csv', index=False)
tas_ground
```
