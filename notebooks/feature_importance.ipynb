```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

feature_importance = pd.read_csv("../ICESAT/feature_importance.csv")
feature_importance.columns = ['features', 'importance']

feature_importance['features'] = feature_importance['features'].str.extract(r'(\w+)_')
feature_importance = feature_importance.groupby('features')['importance'].sum().reset_index()


sorted_importance = feature_importance.sort_values(by='importance', ascending=False)
sorted_importance

```

```python
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_importance)), sorted_importance['importance'], align='center')
plt.yticks(range(len(sorted_importance)), sorted_importance['features'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Sorted Feature Importance Plot')
plt.show()
```

```python
plt.barh(y = np.arange(len(feature_importance)), width=feature_importance)

plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance Bar Chart')
plt.show()


# forest_importances = pd.Series(importances, index=feature_names)

# fig, ax = plt.subplots()
# forest_importances.plot.bar(yerr=std, ax=ax)
# ax.set_title("Feature importances using MDI")
# ax.set_ylabel("Mean decrease in impurity")
# fig.tight_layout()
```
