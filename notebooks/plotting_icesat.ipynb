```python
import pandas as pd
import numpy as np 
import rasterio
import rioxarray
import matplotlib.pyplot as plt 
import xarray as xr
import math
import seaborn as sns
sns.set()
sns.set_context("notebook")
```

```python
taz = r"../ICESAT/tasmania/tasmania.csv"

icesat = pd.read_csv(taz)
latitudes = icesat['lat'][:]
longitudes = icesat['lon'][:]
elevations = icesat['h_te_interp'][:]

target_latitude = -42.91 # Center latitude
latitude_tolerance = 0.01 # Tolerance range (+/- 0.2 degrees)

# Find indices within the latitude range
indices = np.where((latitudes >= target_latitude - latitude_tolerance) & (latitudes <= target_latitude + latitude_tolerance))
# Extract data for the selected latitude range
selected_longitudes = longitudes[indices[0]]
selected_elevations = elevations[indices[0]]

plt.figure(figsize=(10, 6))
plt.scatter(selected_longitudes, selected_elevations)
plt.xlabel('Latitude')
plt.ylabel('Elevation (meters)')
plt.title('ICESat-2 Cross-Section Data')
plt.grid(True)
plt.show()

```
